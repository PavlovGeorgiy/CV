```python
# При помощи SQL и Python получите датафрейм с названием города и его населением из таблицы City
gor = pd.read_sql(
"""
select
  name, population
from
  City
WHERE
 name = 'Antalya' or
 name = 'Soti' or
 name = 'Pak Kret' or
 name = 'Alanya' or
 name = 'Nicosia' or
 name = 'Abu Dhabi' or
 name = 'Male' or
 name = 'Hanoi' or
 name = 'Roma'
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
gor.head(10)
```

```python
# При помощи Python соедините данные из выгруженного CSV файла (пункт 8) и таблицы с населением города
last = itog.merge(gor, left_on='city_eng', right_on='name', how='inner')
last.head(10)
```

```python
# Сгруппируйте итоговый датафрейм по странам
last.groupby('страна')
last[['страна', 'город', 'population','начало отпуска','окончание отпуска', 'планируемый бюджет', 'стоимость билетов',
'стоимость отеля', 'отель за день', 'траты на отдыхе','траты на месте в день','общая оценка отпуска', 'итоговая стоимость отпуска',
'отклонение от плана','стоимость 1 дня отпуска',]]
```
